```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
df = pd.read_csv('/content/toyota_sustainability_full.csv')

# Display basic info
print(df.info())

```

```python
# Display first few rows
df.head()
```

```python
df.describe()
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Sustainability Metrics")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Line plot for sustainability trends
sns.lineplot(x=df['Year'], y=df['GHG Emissions (Metric Tons)'], label='GHG Emissions')
sns.lineplot(x=df['Year'], y=df['Water Usage (Million Liters)'], label='Water Usage')
sns.lineplot(x=df['Year'], y=df['Energy Consumption (GWh)'], label='Energy Consumption')

plt.xlabel("Year")
plt.ylabel("Values")
plt.title("Sustainability Trends Over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prepare data
df_grouped = df.groupby("Year").mean().reset_index()
X = df_grouped[['Year']].values
y = df_grouped['GHG Emissions (Metric Tons)'].values

# Train regression model
model = LinearRegression()
model.fit(X, y)

# Predict future emissions
future_years = np.array([[2024], [2025], [2026]])
predictions = model.predict(future_years)

# Display predictions
for year, pred in zip(future_years.flatten(), predictions):
    print(f"Predicted GHG Emissions for {year}: {pred:.2f}")

# Plot predictions
plt.figure(figsize=(8, 6))
plt.scatter(df_grouped['Year'], df_grouped['GHG Emissions (Metric Tons)'], color='blue', label='Actual Data')
plt.plot(future_years, predictions, color='red', linestyle='dashed', marker='o', label='Predicted Data')
plt.xlabel('Year')
plt.ylabel('GHG Emissions (Metric Tons)')
plt.title('GHG Emissions Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['GHG Emissions (Metric Tons)', 'Water Usage (Million Liters)', 'Energy Consumption (GWh)']])
plt.title('Boxplot for Sustainability Metrics')
plt.show()

```

```python
# Bar Chart: Multiple Sustainability Metrics Over the Years
metrics = ["CO2 Emissions Reduction (%)", "Renewable Energy Usage (%)", "Water Efficiency Improvement (%)", "Carbon Neutrality Progress (%)"]
df_grouped = df.groupby("Year")[metrics].mean()

df_grouped.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.title("Sustainability Metrics Over the Years")
plt.legend(title="Metrics")
plt.xticks(rotation=0)
plt.show()
```

```python
 #Comparison of All Metrics Over the Years
plt.figure(figsize=(12, 6))
for metric in metrics:
    sns.lineplot(x=df['Year'], y=df[metric], label=metric)
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.title("Comparison of Sustainability Metrics Over the Years")
plt.legend()
plt.show()
```

```python
# Pie Chart: Composition of Raw vs. Recycled Materials
labels = ['Recycled Materials', 'Raw Materials Used']
sizes = [df['Recycled Materials Use Rate (%)'].mean(), 100 - df['Recycled Materials Use Rate (%)'].mean()]
colors = ['green', 'gray']
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Proportion of Recycled vs. Raw Materials Usage")
plt.show()
```

```python
# Scatter Plot: Relationship between Energy Consumption & CO2 Reduction
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Energy Consumption (GWh)'], y=df['CO2 Emissions Reduction (%)'], alpha=0.5)
plt.xlabel("Energy Consumption (GWh)")
plt.ylabel("CO2 Emissions Reduction (%)")
plt.title("Energy Consumption vs. CO2 Reduction")
plt.show()
```

```python
# Line Chart: Renewable Energy Usage
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Year'], y=df['Renewable Energy Usage (%)'], ci=None, marker='o', color='blue')
plt.xlabel("Year")
plt.ylabel("Renewable Energy Usage (%)")
plt.title("Trend of Renewable Energy Usage")
plt.show()
```
